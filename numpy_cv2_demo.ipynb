```python
#random 변수 생성
np.random.seed(0)
array = np.random.randint(0,10,(2,3))
print(array)
```

```python
#numpy 객체 복사
array1 = np.arange(0,10)
array2 = array1
array2[0] = 99
print(array1[0],array2[0])
```

```python
array3 = np.arange(0,10)
print(array3)
array4 = array3.copy()
print(array4,array3)
array4[0] = 99
print(array3, array4)
print(array3[0],array4[0])
```

```python

list1 = [1,2,3,4]
list2 = list1[:]
list2[0] = 99
print(list1[0], list2[0])
```

```python
# 중복된 원소 제거

array = np.array([2,1,3,4,1,2,3,1,3,5])
print(f"중복처리전  : {array}")
print(f"중복처리후 : {np.unique(array)}")
```

```python
#np.isin() ->  내가 찾는 원소가 있는지 index와 bool값으로 리턴
array = [1,2,3,7,8,11]
array1 = np.array([1, 2, 3, 4])
print(np.isin(array1, array))
```

```python
import matplotlib.pyplot as plt
image_path = "./face_recognition.JPG"

image = cv2.imread(image_path)
#rgb변환
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_50x50 = cv2.resize(image_rgb,(50,50))

flg, ax = plt.subplots(1,2, figsize=(10,5))
print(flg)
ax[0].imshow(image_rgb)
ax[0].set_title("Original Image")

ax[1].imshow(image_50x50)
ax[1].set_title("Resized Image")

```

```python
#저장
cv2.imwrite("./iamge_50x50.png",image_50x50)
```

```python
def image_show(image):
  cv2_imshow(image)
  cv2.waitKey(0)

image_path = "./face_recognition.JPG"

#이미지읽기

image = cv2.imread(image_path)

image_crop = image[50:, :500]

image_show(image_crop)

cv2.imwrite("./image_crop.png", image_crop)
```

```python
def imshow(image):
  cv2_imshow(image)
  cv2.waitKey(0)

image_path = "./face_recognition.JPG"

image = cv2.imread(image_path)

image_blur = cv2.blur(image,(3,3))
imshow(image_blur)
```
