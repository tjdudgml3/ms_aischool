```python
import matplotlib.pyplot as plt
```

```python
#Style적용하기
plt.style.use('classic')
```

```python
import numpy as np
```

```python
#Sample Data 만들기
x = np.linspace(0,10,100)
x
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.cos(x))
plt.plot(x,np.sin(x))
```

```python
fig = plt.figure()
plt.plot(x,np.cos(x),'-')
# plt.figure()
plt.plot(x,np.sin(x),'--')
fig.savefig('my_figure.png')
```

```python
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.subplot(2,2,3)
plt.plot(x,np.sin(x))
plt.subplot(2,2,4)
plt.plot(x,np.cos(x))
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,100)
ax.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x-0))
plt.plot(x,np.sin(x-1),color = 'b')
plt.plot(x,np.sin(x-2),color = 'g')
plt.plot(x,np.sin(x-3),color = 'w')
plt.plot(x,np.sin(x-4),color = 'k',linestyle ='dashed')
plt.plot(x,np.sin(x-5),color = '#FF00FF')
```

```python
plt.plot(x,x+0,'-g')
plt.plot(x,x+1,'-.g')
plt.plot(x,x+2,':c')
plt.plot(x,x+3,'--g')
```

```python
#플롯 조정하기 : 축 경계
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
```

```python
#플롯에 레이블 붙이기
plt.plot(x,np.sin(x))
plt.title('A Sin Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x,np.sin(x),'-g',label = 'sin(x)')
plt.plot(x,np.cos(x),'--b', label = 'cos(x)')
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('A sin curve')
plt.legend()
```

```python
x= np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'o')
```

```python
rng = np.random.RandomState(0)
for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5),rng.rand(5),marker,label = 'marker = {0}'.format(marker))
plt.legend()
plt.xlim(0,1.8)
```
